```python
def count_outliers_iqr(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = column[(column < lower_bound) | (column > upper_bound)]
    return len(outliers), lower_bound, upper_bound

# Check outliers for all numerical columns
numerical_cols = ['stars', 'reviews', 'price', 'category_id']

print("=== NUMBER OF OUTLIERS USING IQR METHOD ===")
for col in numerical_cols:
    outlier_count, lower, upper = count_outliers_iqr(df[col])
    percentage = (outlier_count / len(df)) * 100
    print(f"{col}: {outlier_count} outliers ({percentage:.2f}%)")
    print(f"   Range: [{lower:.2f}, {upper:.2f}]")
```

```python
# attribute:https://www.geeksforgeeks.org/machine-learning/derivative-of-the-sigmoid-function/
# author: geeksforgeeks

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# attribute:https://www.datacamp.com/tutorial/the-cross-entropy-loss-function-in-machine-learning
# author: Kurtis Pykes
def loss(y_true, y_pred):
    """
    y_true = target variable
    y_pred = predictions.
    """ 
    # binary cross entropy
    epsilon = 1e-9
    y1 = y_true * np.log(y_pred + epsilon)
    y2 = (1-y_true) * np.log(1 - y_pred + epsilon)
    loss = -np.mean(y1 + y2)
    return loss
```

```python
# attribute:https://www.datacamp.com/tutorial/the-cross-entropy-loss-function-in-machine-learning
# author: Kurtis Pykes
#The loss function you provided is the binary cross-entropy loss function, which is widely used in binary classification tasks in machine learning. It measures the dissimilarity between the true labels (y true ) and the predicted probabilities (ypred ) produced by a model, such as a neural network.
#For each sample, if the true label yi is 1, the loss depends heavily on log (ypredi)log); the closer ypredi is to 1, the lower the loss.
```

```python
def gradients(X, y, y_hat):
    """
    X = Input variables
    y = target variable
    y_hat = predictions.
    """    
    # w = weights (parameter).
    # b = bias (parameter).
    
    # n_samples-> number of training examples.
    n_samples, n_features = X.shape
    
    # Gradient of loss (weights).
    dw = (1/n_samples) * np.dot(X.T, (y_hat - y))
    
    # Gradient of loss (bias).
    db = (1/n_samples)*np.sum((y_hat - y))

    return dw, db
```
